```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["WWE"])
data = trends.interest_by_region()
data = data.sort_values(by="WWE", ascending=False)
data = data.head(25)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="WWE", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=720)
data.build_payload(kw_list=['WWE'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 10))
data['WWE'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for WWE', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
